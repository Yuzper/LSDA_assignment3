```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
from matplotlib import pyplot as plt
import numpy as np
import mlflow
```

```python
num_samples = 200
np.random.seed(42)
def get_ys(xs):
    signal = -0.1*xs**3+xs**2 - 5*xs - 5
    noise = np.random.normal(0,200,(len(xs),1))
    return signal + noise

X = np.random.uniform(-20,20,num_samples).reshape((-1,1))
y = get_ys(X)

plt.scatter(X,y)
```

# Polynomial Model

```python
def model_pipeline(degree_):
    model = Pipeline([
        ("Poly 2", PolynomialFeatures(degree = degree_)),
        ("LinReg", LinearRegression())
    ])
    model.fit(X,y)
    
    plotting_x = np.linspace(-20,20,num=50).reshape((50,1))
    r2 = r2_score(y,model.predict(X))
    plotting_preds = model.predict(plotting_x)
    plt.plot(plotting_x,plotting_preds, label=f"degree={i}")
    print("r2: %.2f" % r2)
    
    plt.scatter(X,y)
    plt.show()

    
for i in range(1,6):
    print("Degree:", i)
    model_pipeline(i)
    print()
```

# MLFlow

```python
for degree in range(1,6):
    mlflow.start_run()
    # Logging params
    mlflow.log_params({
        "num_samples": num_samples,
        "poly_degree": degree
    })
    
    model = Pipeline([
        ("Poly", PolynomialFeatures(degree=degree)),
        ("LinReg",LinearRegression())
    ])
        
    model.fit(X,y)
        
    # Logging Metrics
    r2 = r2_score(y, model.predict(X))
    mlflow.log_metric("r2", r2)
        
    # Logging Artifacts
    plt.clf()
    plotting_x = np.linspace(-20,20,num = 50).reshape((50,1))
    plotting_preds = model.predict(plotting_x)
    plt.scatter(X,y,label="data")
    plt.plot(plotting_x, plotting_preds, label=f"degree")
    plt.legend()
    filename = "plots/PolyReg_%d_%d.png" % (degree, num_samples)
    plt.savefig(filename)
    mlflow.log_artifact(filename)
        
    # Finally log the model
    mlflow.sklearn.log_model(model, "model")
    mlflow.end_run()
    
    
```
